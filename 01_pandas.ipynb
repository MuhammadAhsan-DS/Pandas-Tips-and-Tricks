# Assignment: Mean, Median, and Mode

## 1️⃣ Measures of Central Tendency

| Term     | Definition                       | Example                  |
|----------|---------------------------------|--------------------------|
| **Mean** | Average of all numbers           | (2 + 3 + 5) ÷ 3 = 3.33  |
| **Median** | Middle value when sorted        | [2, 3, 5] → Median = 3  |
| **Mode** | Most frequent value             | [2, 3, 3, 5] → Mode = 3 |

**Use in Missing Values:**
- **Mean** → numeric, normally distributed  
- **Median** → numeric, skewed  
- **Mode** → categorical
