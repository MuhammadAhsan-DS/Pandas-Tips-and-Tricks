# Automatic EDA using `Ydata_Profiling`

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import ydata_profiling as yd
```

```python
# import data from seaborn
df = sns.load_dataset("titanic")
```

### Automatic EDA Titanic Dataset

```python
# ydata profiling report
report = yd.ProfileReport(df)
profile = report.to_file(output_file="./Outputs/titanic_report.html")
```
